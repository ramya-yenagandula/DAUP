```python
import pandas as pd
df = pd.read_csv("/content/sample_data/california_housing_train.csv")

numerical_cols = df.select_dtypes(include=['number'])

stats = {
    "Mean": numerical_cols.mean(),
    "Median": numerical_cols.median(),
    "Standard Deviation": numerical_cols.std(),
    "Interquartile Range (IQR)": numerical_cols.quantile(0.75) - numerical_cols.quantile(0.25),
    "Skewness": numerical_cols.skew(),
    "Kurtosis": numerical_cols.kurt()
}

stats_df = pd.DataFrame(stats)

print(stats_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
numerical_cols = df.select_dtypes(include=['number'])
cols_to_plot = ['households', 'median_house_value']
plt.figure(figsize=(6, 6))
sns.boxplot(data=df[cols_to_plot])
plt.title('Box Plot of Housing Median Age and Median House Value')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```
